```python
def find_unique_tags(df): 
    uni_tags = []
    total_tags = 0

    for t in tag_col:
        unique_tags = df[t].unique()
        #print(t + " has unique tags of " + str(len(unique_tags)))
        uni_tags.append(list(unique_tags))

    total_uni_tags = sum(uni_tags, [])
    total_uni_tags = set(total_uni_tags)
    total_uni_tags = total_uni_tags.difference([1.0, ' '])
    return total_uni_tags
```

```python
def create_tag_dict(df):
    uni_tags_dict = {}
    total_uni_tags = find_unique_tags(df)
    for tag in tag_col:
        for t in df[tag]:
            if t in total_uni_tags:
                uni_tags_dict[t] = uni_tags_dict.get(t,0) + 1
                
    return uni_tags_dict
```

```python
# what are the least popular issue democracts/republican/independent addressed? 
def party_issue_tags(df):
    party_issue = {}
    ts = find_unique_tags(df)
    for p in df["Party"].unique():
        party_issue[p] = {}
        for t in ts:
            party_issue[p][t] = 0
    for index, row in df.iterrows():
            p = row["Party"]
            for t in row[tag_col]:
                try:
                    party_issue[p][t] = party_issue[p][t] + 1
                except KeyError:
                    pass
    return party_issue
```

```python
issue_dict = create_tag_dict(df)
```
