```python
import os
import csv
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from operator import itemgetter
```

```python
df = pd.read_csv("joined_stances.csv")
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df
```

```python
# Remove stances
# Reform
# Government operations
# Diversity / Discrimination
# Techology
# Local
# Seniors
remove_issues = ["Reform", "Government Operations", "Diversity / Discrimination", "Technology", "Local", "Seniors"]
for r in remove_issues:
    df = df[df["Issue Tag 1"] != r]
```

```python
df["Issue Tag 1"] = df["Issue Tag 1"].str.strip()
```

```python
# Combine stances 
# “Taxes”, combine with  "Taxes / Budget” -- becomes “Taxes / Budget” 
# “Support Services” and “Social Services -- becomes “Social Services / Support Services”
# “Transparency + Participation” and “Citizen Participation -- becomes “Transparency / Participation”
# “Healthcare” and “Healthcare Reform” and “Public Health” -- becomes “Healthcare”
# “Criminal Justice” and “Public Safety + Criminal Justice” and “Court/Prison System”-- “Public Safety / Criminal Justice”
# “Transportation” and “Infrastructure” and “Public Transit” -- “Transportation / Infrastructure”
# “Environment” and “Energy” -- “Environment / Energy”
# “Marriage” and “LGBTQ” -- “Marriage / LGBTQ”
# "Civil Liberties" -- "Civil Liberties / Privacy"
# "Privacy" -- "Civil Liberties / Privacy"

df["Issue Tag 1"].replace({"Taxes":"Taxes / Budget", "Support Services": "Social Services / Support Services",
                                        "Social Services": "Social Services / Support Services", 
                                        "Transparency + Participation":"Transparency / Participation",
                                        "Citizen Participation":"Transparency / Participation",
                                        "Healthcare Reform":"Healthcare",
                                        "Public Health":"Healthcare",
                                        "Criminal Justice": "Public Safety / Criminal Justice",
                                        "Public Safety + Criminal Justice": "Public Safety / Criminal Justice",
                                        "Court/Prison System": "Public Safety / Criminal Justice",
                                        "Transportation": "Transportation / Infrastructure",
                                        "Infrastructure": "Transportation / Infrastructure",
                                        "Public Transit": "Transportation / Infrastructure",
                                        "Environment":"Environment / Energy",
                                        "Energy": "Environment / Energy",
                                        "Marriage": "Marriage / LGBTQ",
                                        "LGBTQ": "Marriage / LGBTQ",
                           "Budget": "Taxes / Budget",
                           "Civil Liberties": "Civil Liberties / Privacy",
                           "Privacy": "Civil Liberties / Privacy"
                                    }, inplace=True)
```
